```python
# Start by importing
import matplotlib.pyplot as plt

# Will also need numpy and pandas for data
import numpy as np
import pandas as pd
```

```python
#Importing the csv from local folder
stocks = pd.read_csv('stocks.csv')
#A quick look at whats come in
stocks.info
```

```python
#Select the start of trading from date column, develope a definition/function
```

```python
#Find the starting value for late entries into market, we will use as marker in chart
def find_start_date(target1,target2):
    for i in range(8364):
        if target1[i]>0:
            #print(target2[i])
            ans = target2[i]
            break
    return ans
            
```

```python
#Load the columns into variables from the import data frame
```

```python
x = stocks["Date"]
y1 =stocks["Apple"]
y2 =stocks["Google"]
y3 =stocks["Facebook"]
y4 =stocks["Amazon"]
y5 =stocks["Netflix"]
y6 =stocks["Microsoft"]
```

```python
# The plot function gives us line plots. Good thing about the object oriented programming method

#Using modulus to indentify tick mark positions, (12,657) are suitable as a divisor pair
collect = [0]
for i in range(697,8364):
    count = i+1
    if count % 697 == 0:
        collect.append(stocks["Date"][i])

# Form the plot figure and border size
fig, ax = plt.subplots( figsize=(20, 16))

ax.set_xticks(np.arange(0,8364,697), minor=False)
ax.set_xticklabels(collect) #From modulas calc above

#Choose lines to plot and their style
line1 = ax.plot(x,y1, label='Apple', color = "purple", lw=3,ls='solid')
line2 = ax.plot(x,y2, label='Google', color = "pink", lw =3,ls='dashed')
line3 = ax.plot(x,y3, label='Facebook', color = "blue", lw=3,ls='dashdot')
line4 = ax.plot(x,y4, label='Amazon', color = "green", lw=3,ls='dotted')
line5 = ax.plot(x,y5, label='Netflix', color = "black", lw=3,ls='--')
line6 = ax.plot(x,y6, label='Microsoft', color = "red", lw=3,ls='-.')
#Show on the plot when theses business started up
ax.annotate('Apple startup',
            xy=('07/09/1984', 10), xycoords='data',
            xytext=(0.1, 0.1), textcoords='axes fraction',
            arrowprops=dict(facecolor='purple', shrink=0.02),
            horizontalalignment='right', verticalalignment='top')
ax.annotate('Microsoft startup',
            xy=('12/03/1986', 60), xycoords='data',
            xytext=(0.1, 0.3), textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.02),
            horizontalalignment='right', verticalalignment='top')
ax.annotate('Amazon startup',
            xy=('20/05/1997', 5), xycoords='data',
            xytext=(0.3, 0.2), textcoords='axes fraction',
            arrowprops=dict(facecolor='green', shrink=0.02),
            horizontalalignment='right', verticalalignment='top')
ax.annotate('Google startup',
            xy=('19/08/2004', 20), xycoords='data',
            xytext=(0.5, 0.3), textcoords='axes fraction',
            arrowprops=dict(facecolor='pink', shrink=0.02),
            horizontalalignment='right', verticalalignment='top')
ax.annotate('Netflix startup',
            xy=('28/02/2005', 10), xycoords='data',
            xytext=(0.62, 0.27), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.01),
            horizontalalignment='right', verticalalignment='top')
ax.annotate('Facebook startup',
            xy=('18/05/2012', 100), xycoords='data',
            xytext=(0.80, 0.175), textcoords='axes fraction',
            arrowprops=dict(facecolor='blue', shrink=0.001),
            horizontalalignment='left', verticalalignment='top')

# Add a legend using the labels defined above
ax.legend()

plt.show()
```
